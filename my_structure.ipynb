```python
#5
class Car:
    pass

cars = []

дані = [
    ("Toyota Corolla" , 2018, 38000),
    ("BMW X5", 2020, 75000),
    ("Renault Duster", 2019, 35000),
    ("Hyundai Elantra", 2021, 42000),
    ("Skoda Octavia", 2017, 39000)
]

for марка, рік, ціна in дані:
    c = Car()
    c.марка = марка
    c.рік = рік
    c.ціна = ціна
    cars.append(c)

print("Автомобілі на складі:")
for car in cars:
    print(f"Марка: {car.марка}, Рік: {car.рік}, Ціна: {car.ціна} грн")

print("\nАвтомобілі з ціною менше ніж 40000 грн:")
for car in cars:
    if car.ціна < 40000:
        print(f"{car.марка}, {car.рік}")
```
