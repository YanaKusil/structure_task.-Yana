```python
#5
class Student:
    pass

students = {}
n = int(input("Скільки студентів ви хочете додати? "))

for i in range(n):
    student = Student()
    student.name = input("Введіть ПІП студента: ")
    student.course = int(input("Введіть курс: "))
    student.group = input("Введіть номер групи: ")
    student.address = input("Введіть місце проживання: ")
    students[i] = student

print("\nСтуденти, які живуть у гуртожитку:")
for i in students:
    student = students[i]
    if student.address == "гуртожиток":
        print(f"{student.name}, курс: {student.course}, група: {student.group}")
```
